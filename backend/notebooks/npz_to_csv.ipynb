# Convert .npz files to CSV

- Reads first 500 augmented `.npz` files from each folder in `keypoints_augmented/`
- Converts each to a `.csv` file in the same folder

```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
MODELS = [
    Path('../data/processed/keypoints_augmented/movenet'),
    Path('../data/processed/keypoints_augmented/mediapipe_norm'),
    Path('../data/processed/keypoints_augmented/mediapipe_world'),
]
```

```python
LIMIT = 500

for model_dir in MODELS:
    npz_files = sorted(model_dir.glob('*.npz'))[:LIMIT]

    for npz_file in npz_files:
        f = np.load(npz_file)
        df = pd.DataFrame(f['data'], columns=f['columns'])
        df.to_csv(npz_file.with_suffix('.csv'), index=False)

    print(f'{model_dir.name}: {len(npz_files)} files converted')
```
